```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# y variable

data_counts = df['blueWins'].value_counts()
colors = ['lightgreen','lightblue']
plt.pie(data_counts, labels=data_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title(f'Pie Chart of blueWins')
plt.axis('equal')
plt.show()
```

```python
features_1 = ['blueFirstBlood','blueEliteMonsters','blueDragons','blueHeralds','blueTowersDestroyed',
              'redFirstBlood','redEliteMonsters','redDragons','redHeralds','redTowersDestroyed']

for feature in features_1:
    data_counts = df[feature].value_counts()
    colors = ['lightgreen','lightblue','lightgrey']
    plt.pie(data_counts, labels=data_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
    plt.title(f'Pie Chart of {feature}')
    plt.axis('equal')
    plt.show()
```

```python
features_2 = ['blueWardsPlaced','blueWardsDestroyed','blueKills','blueDeaths',
            'blueAssists','blueTotalGold','blueAvgLevel','blueTotalExperience',
            'blueTotalMinionsKilled','blueTotalJungleMinionsKilled','blueGoldDiff',
            'blueExperienceDiff','blueCSPerMin','blueGoldPerMin','redWardsPlaced',
            'redWardsDestroyed','redKills','redDeaths','redAssists','redTotalGold',
            'redAvgLevel', 'redTotalExperience','redTotalMinionsKilled','redTotalJungleMinionsKilled',
            'redGoldDiff','redExperienceDiff','redCSPerMin','redGoldPerMin']

for feature in features_2:
    plt.hist(df[feature], bins=20, color='lightblue', edgecolor='white')
    plt.title(f'Data Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
    
    max=df[feature].max()
    min=df[feature].min()
    print(f'The max of {feature} is: {max}')
    print(f'The min of {feature} is: {min}')
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
